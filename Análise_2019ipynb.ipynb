```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cre_2019 = pd.read_excel('2019 agrupado.xlsx')
cre_2019.head()
```

```python
cre_2019['InstituicaoFinanceira'].unique()
```

```python
cre_2019['Segmento'].unique()
```

```python
cre_2019['Modalidade'].unique()
```

```python
media = cre_2019.groupby(by='Mês').mean().reset_index()
media = media.sort_values(by='Mês', ascending=False)
media["Mês"] = media["Mês"].replace({
    "Janeiro": 1,"Fevereiro": 2,"Março": 3,"Abril": 4,"Maio": 5,"Junho": 6,'Julho': 7,
    'Agosto':8,'Setembro':9,'Outubro':10,'Novembro':11,'Dezembro':12
})
media.sort_values(by='Mês',ascending=False)
media
```

```python
media.info()
```

```python
plt.title('Média de taxa ao mês do ano de 2019',fontsize=14)
sns.lineplot(x='Mês',y='TaxaJurosAoMes',data=media)
plt.ylabel('Taxa')
plt.xlabel('Mês');
```

```python
plt.title('Média de taxa ao ano do ano de 2019',fontsize=14)
sns.lineplot(x='Mês',y='TaxaJurosAoAno',data=media)
plt.ylabel('Taxa')
plt.xlabel('Mês');
```

```python
sns.boxplot(y='TaxaJurosAoMes',x='Mês',data=cre_2019,color='royalblue')
plt.xticks(rotation=90);
```

```python
sns.boxplot(y='TaxaJurosAoAno',x='Mês',data=cre_2019,color='royalblue')
plt.xticks(rotation=90);
```
